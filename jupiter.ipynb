```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
#creating dataset
data = {
    'Study_Hours': [1,2,3,8,9,10,2,3,9,10],
    'Attendance': [40,45,50,85,90,95,42,48,88,96]
}

df = pd.DataFrame(data)

df
```

```python
#selecting features
X = df[['Study_Hours', 'Attendance']]
```

```python
#creating models K = Number of clusters

#Ex:K=2 → two groups, K=3 → three groups
kmeans = KMeans(
    n_clusters=2,
    random_state=42
)
```

```python
#training the models
kmeans.fit(X)
```

```python
#Adding cluster labels
df['Cluster'] = kmeans.labels_
df
```

```python
#plotting clusters
plt.scatter(
    df['Study_Hours'],
    df['Attendance'],
    c=df['Cluster']
)

plt.xlabel("Study Hours")
plt.ylabel("Attendance")
plt.title("K-Means Clustering")

plt.show()
```

```python
#plotting cluster center
plt.scatter(
    df['Study_Hours'],
    df['Attendance'],
    c=df['Cluster']
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=200,
    marker='X'
)

plt.xlabel("Study Hours")
plt.ylabel("Attendance")

plt.show()
```
